```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


# Data for different charts
technologies = ['AI-based tools', 'Drones', 'IoT devices', 'Remote sensing', 'None of these']
tech_responses = [7, 3, 2, 4, 21]

# Pie chart data for AI Familiarity
labels = ['Very Familiar', 'Somewhat Familiar', 'Not Familiar']
ai_familiarity = [8, 22, 7]

# Stacked bar chart data for AI Benefits Areas
areas = ['Crop health', 'Pest management', 'Irrigation', 'Soil health', 'All Options']
benefits = np.array([[7, 9, 5, 7, 1]])

# Horizontal bar chart data for Experience with Digital Platforms
experience = ['Extensive', 'Some', 'Minimal', 'None']
experience_counts = [8, 15, 8, 6]

# Line graph data for AI Usage Interest
stages = ['Interested', 'Not Interested', 'Undecided']
usage_interest = [29, 2, 6]

# Heatmap data for Technology Usage
heatmap_data = np.array([[10, 15, 7, 0], [5, 12, 8, 10]])

# --- Bar Chart ---
plt.figure(figsize=(12, 8))
plt.subplot(2, 3, 1)
plt.bar(technologies, tech_responses, color='skyblue')
plt.title('Technology Usage in Agriculture')
plt.xlabel('Technology')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45, ha='right')

# --- Pie Chart ---
plt.subplot(2, 3, 2)
plt.pie(ai_familiarity, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Familiarity with AI in Agriculture')
plt.axis('equal')

# --- Stacked Bar Chart ---
plt.subplot(2, 3, 3)
plt.bar(areas, benefits[0], color=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])
plt.title('AI Benefits in Farming Areas')
plt.xlabel('Areas of Farming')
plt.ylabel('Number of Responses')

# --- Horizontal Bar Chart ---
plt.subplot(2, 3, 4)
plt.barh(experience, experience_counts, color='lightseagreen')
plt.title('Experience with Digital Platforms')
plt.xlabel('Number of Responses')
plt.ylabel('Experience Level')

# --- Line Graph ---
plt.subplot(2, 3, 5)
plt.plot(stages, usage_interest, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title('AI Usage Interest')
plt.xlabel('Stages')
plt.ylabel('Number of Responses')
plt.grid(True)

# --- Heatmap ---
plt.subplot(2, 3, 6)
sns.heatmap(heatmap_data, annot=True, cmap='Blues', xticklabels=['AI-based tools', 'Drones', 'IoT', 'Others'],
            yticklabels=['Positive', 'Negative'])
plt.title('Heatmap of Responses to Technology Usage')
plt.xlabel('Technology')
plt.ylabel('Response Type')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
categories = [
    "Familiarity with AI",
    "Current Use of Technologies",
    "Willingness to Adopt AI",
    "Experience with Digital Platforms",
    "Perception of AI Recommendations",
    "Barriers to Adoption",
    "Interest in Workshops",
    "Role in Climate Resilience",
    "Use of Farm Management Apps",
    "Comfort with Drones"
]

values = [33.3, 30, 78.4, 25, 24.3, 48.6, 75.7, 40.5, 25, 50]

# Set style and color palette
sns.set_theme(style="whitegrid")
colors = sns.color_palette("viridis", len(categories))

# Create the bar chart
plt.figure(figsize=(12, 7))
bars = sns.barplot(x=values, y=categories, palette=colors)

# Add percentage labels to each bar
for index, value in enumerate(values):
    plt.text(value + 1, index, f'{value:.1f}%', va='center', fontsize=10, color='black')

# Chart aesthetics
plt.title("Insights on Technology Use in Sustainable Agriculture", fontsize=16, pad=20)
plt.xlabel("Percentage (%)", fontsize=12)
plt.ylabel("")  # No label for y-axis for clean design
plt.xlim(0, 100)  # Ensure consistent scale
plt.tight_layout()

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors

# Data
categories = [
    "Willingness to adopt AI",
    "Interest in training programs",
    "Market demand for AI insights",
    "Potential role in climate resilience",
    "Lack of familiarity with AI",
    "Barriers to adoption",
    "Limited comfort with drones/IoT",
    "Minimal experience with digital platforms"
]
values = [78.4, 75.7, 45.9, 40.5, -59.5, -48.6, -43.2, -40.5]

# Generate color shades
positive_colors = plt.cm.Greens(np.linspace(0.4, 1, len([v for v in values if v > 0])))
negative_colors = plt.cm.Reds(np.linspace(0.4, 1, len([v for v in values if v < 0])))

colors = []
pos_idx, neg_idx = 0, 0
for val in values:
    if val > 0:
        colors.append(positive_colors[pos_idx])
        pos_idx += 1
    else:
        colors.append(negative_colors[neg_idx])
        neg_idx += 1

# Create the horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bars = ax.barh(categories, values, color=colors, edgecolor='black')

# Add data labels
for bar in bars:
    width = bar.get_width()
    ax.text(
        width + (2 if width > 0 else -2),  # Adjust text position
        bar.get_y() + bar.get_height() / 2,
        f"{abs(width):.1f}%",
        ha='left' if width > 0 else 'right',
        va='center',
        color='black'
    )

# Title and labels
ax.set_title("Positive and Negative Insights in Sustainable Agriculture", fontsize=14)
ax.set_xlabel("Percentage of Respondents", fontsize=12)
ax.set_ylabel("Insights", fontsize=12)
ax.axvline(0, color='black', linewidth=0.8)  # Add a vertical line at 0 for clarity
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for visualization
categories = ['Willingness to Adopt AI', 'Interest in Training', 'Demand for AI Insights',
              'Climate Resilience Potential', 'Lack of Familiarity with AI',
              'Cost and Trust Issues', 'Comfort with Advanced Tech', 'Experience with Digital Tools']

values = [78.4, 75.7, 45.9, 40.5, -59.5, -16.2, -43.2, -40.5]  # Positive (positive values) and Negative (negative values)

colors = ['#4CAF50', '#8BC34A', '#CDDC39', '#FFEB3B', '#FF7043', '#FF5722', '#E64A19', '#D32F2F']  # Positive: Greenish | Negative: Reddish
explode = [0.1 if value > 0 else 0 for value in values]  # Highlight positive categories

# Plot
plt.figure(figsize=(10, 6))
plt.pie(
    [abs(value) for value in values], labels=categories, autopct='%1.1f%%', startangle=140,
    colors=colors, explode=explode
)
plt.title('Positive vs Negative Factors in Technology Adoption')
plt.show()
```
